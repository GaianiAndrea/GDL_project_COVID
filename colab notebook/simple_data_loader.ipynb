```python
simple_2020 = final_2020[["Date", "FIPS", "Cases"]]
simple_2020.head()
```

```python
date = simple_2020.Date.unique()
# print(date)
```

```python
fips_date_cases_2020 = {}

for index, row in simple_2020.iterrows():
  if (row['Date'],row['FIPS']) not in fips_date_cases_2020:
    fips_date_cases_2020[row['Date'], row['FIPS']] = row['Cases']
  else:
    fips_date_cases_2020[row['Date'], row['FIPS']].append(row['Cases'])
```
