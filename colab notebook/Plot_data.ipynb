```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t_loss = pd.read_csv('train_loss.csv')
v_loss = pd.read_csv('val_loss.csv')
t_loss.rename(columns = {'Value':'Train'}, inplace = True)
t_loss['Validation'] = v_loss['Value']
t_loss = t_loss[['Step', 'Train', 'Validation']]
t_loss.drop(t_loss.index[:5], axis=0, inplace=True)
t_loss.drop(t_loss.index[300:999], axis=0, inplace=True)
t_loss.set_index('Step', inplace=True)

t_loss.plot.line(figsize=(10,5))
plt.title('Loss')
plt.legend()
plt.show()
```
